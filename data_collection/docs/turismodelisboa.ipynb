- Since the information that will make up the **RAG system** regarding the creation of itineraries in Lisbon related to tourism is in a `PDF` document, we have to convert it into a data type compatible with the database that will be used.
  - The document is available at: [Turismo de Lisboa](https://www.discoverportugal.info/reisefuehrer/wp-content/uploads//lisboacard_detail.pdf)

```python
# Install the required packages
# !pip install pandas marker-pdf 

# Assume that PyTorch and other necessary libraries are already installed
```

```python
# Import necessary libraries
import pandas as pd
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered
```

```python
# Use PdfConverter to extract text and images from the PDF
# Source: https://github.com/datalab-to/marker?tab=readme-ov-file#use-from-python
converter = PdfConverter(
    artifact_dict=create_model_dict(),
)
rendered = converter("../docs/c755d1d11640ebd527b62ce6490a02ec-Guia_LxCard_Ing_Esp_Abril_2024 (1).pdf")
text, _, images = text_from_rendered(rendered)
```

```python
# Note: Assuming 'text' is a string containing the extracted text
#       and 'images' is a dictionary where keys are page indices (int) 
#       and values are lists of PIL.Image objects, based on marker-pdf output.

# If the structure differs, adjust accordingly.
from PIL import Image                               # For handling and saving images
import os                                           # For creating directories if needed

# Define output directory for saving files
output_dir = "extracted_content"
os.makedirs(output_dir, exist_ok=True)  # Create the directory if it doesn't exist

# Save the extracted text to a file
text_file_path = os.path.join(output_dir, "extracted_text.txt")
with open(text_file_path, "w", encoding="utf-8") as text_file:
    text_file.write(text)  # Write the text content to the file
print(f"Text saved to: {text_file_path}")

# Save the extracted images
# Iterate through the images dictionary directly, since values are single Image objects (not lists)
for original_filename, img in images.items():
    # Generate a safe filename: replace .jpeg with .jpg for saving (or use .png if preferred)
    # This preserves the original naming convention
    img_filename = original_filename.replace('.jpeg', '.jpg')
    img_file_path = os.path.join(output_dir, img_filename)
    
    # Save the image using PIL's save method (saves as JPEG format)
    img.save(img_file_path, format='JPEG')  # Specify format to ensure it's saved correctly
    print(f"Image saved to: {img_file_path}")
```
